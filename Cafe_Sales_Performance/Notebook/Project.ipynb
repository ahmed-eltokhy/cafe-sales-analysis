```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/dirty_cafe_sales.csv')
df.head(10)
```

```python
df.info()
```

```python
df['Item'].value_counts(dropna=False)
```

```python
df['Item'] = df['Item'].replace(['UNKNOWN','ERROR'], np.nan)
```

```python
df.isnull().sum()
```

```python
missing_rate = df['Item'].isnull().mean() * 100
print(missing_rate,'%')
```

```python
df = df.dropna(subset='Item')
```

```python
df['Quantity'].value_counts(dropna=False)
```

```python
df['Quantity'] = df['Quantity'].replace(['UNKNOWN','ERROR'], np.nan)
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
df['Quantity'].isnull().sum()
```

```python
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce').astype('float')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce').astype('float')
```

```python
menu = df.groupby('Item')['Price Per Unit'].median()
menu
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Item'].map(menu))
```

```python
df.loc[
    df['Quantity'].isna(),
    'Quantity'
] = df['Total Spent'] / df['Price Per Unit']
```

```python
df['Quantity'].isnull().sum()
```

```python
df[df['Quantity'].isna()]
```

```python
missing_rate_quantity = df['Quantity'].isnull().mean() * 100
print(missing_rate_quantity,'%')
```

```python
df = df.dropna(subset='Quantity')
```

```python
df.isnull().sum()
```

```python
df['Price Per Unit'].value_counts(dropna=False)
```

```python
df['Total Spent'].value_counts(dropna=False)
```

```python
df.loc [
    df['Total Spent'].isna(),
    'Total Spent'
] = df['Price Per Unit'] * df['Quantity']
```

```python
df['Total Spent'].isnull().sum()
```

```python
df  = df.drop(columns=['Tptal Spent'])
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Payment Method'] = df['Payment Method'].replace(['UNKNOWN','ERROR'], np.nan)
```

```python
df['Payment Method'].value_counts(dropna=False)
```

```python
missing_rate_payment = df['Payment Method'].isnull().mean() * 100
print(missing_rate_payment,'%')
```

```python
df['Location'].value_counts(dropna=False)
```

```python
df['Location'] = df['Location'].replace(['UNKNOWN','ERROR'], np.nan)
```

```python
df.isnull().sum()
```

```python
missing_rate_location = df['Location'].isnull().mean() * 100
print(missing_rate_location,'%')
```

```python
df['Transaction Date'].value_counts(dropna=False)
```

```python
df['Transaction Date'] = df['Transaction Date'].replace(['UNKNOWN','ERROR'], np.nan)
```

```python
df.isnull().sum()
```

```python
missing_rate_date = df['Transaction Date'].isnull().mean() * 100
print(missing_rate_date,'%')
```

```python
cleaned_cafe_sales = df
cleaned_cafe_sales.to_csv('cleaned_cafe_sales.csv', index=False)
```
